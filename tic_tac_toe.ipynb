Play tic_tac_toe with the computer. Computer's field choice is random.

```python
from random import randrange
board = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

def display(board):
    for i in range(3):
        print('+-----+-----+-----+')
        print('|     |     |     |')
        for j in range(3):
            print(f'|  {board[i][j]}  ', end='')
        print('|')
        print('|     |     |     |')
    print('+-----+-----+-----+')

def playerMove(board):
    while True:
        move = int(input('Player move: '))
        for i, inner in enumerate(board):
            for j, x in enumerate(inner):
                if move == x:
                    board[i][j] = 'O'
                    winner = pickWinner(board)
                    if winner is not None:
                        return winner
                    return

def cpuMove(board):
    while True:
        move = randrange(1, 10)
        for i, inner in enumerate(board):
            for j, x in enumerate(inner):
                if move == x:
                    board[i][j] = 'X'
                    winner = pickWinner(board)
                    if winner is not None:
                        return winner
                    return

def pickWinner(board):
    if board[0][0] == board[0][1] == board[0][2]:
        return board[0][0]
    elif board[1][0] == board[1][1] == board[1][2]:
        return board[1][0]
    elif board[2][0] == board[2][1] == board[2][2]:
        return board[2][0]
    elif board[0][0] == board[1][0] == board[2][0]:
        return board[0][0]
    elif board[0][1] == board[1][1] == board[2][1]:
        return board[0][1]
    elif board[0][2] == board[1][2] == board[2][2]:
        return board[0][2]
    elif board[0][0] == board[1][1] == board[2][2]:
        return board[0][0]
    elif board[0][2] == board[1][1] == board[2][0]:
        return board[0][2]
    return None		

def declareWinner(shape):
    if shape == 'O':
        print('You won!')
    else:
        print('CPU won!')

if __name__ == '__main__':
    i = 0
    cpuMove(board)
    while True:
        display(board)
        winner = playerMove(board)
        if winner:
            display(board)
            declareWinner(winner)
            break
        display(board)
        winner = cpuMove(board)
        if winner:
            display(board)
            declareWinner(winner)
            break
        i += 1
        if i == 4:
            display(board)
            print('nobody won')
            break
```
